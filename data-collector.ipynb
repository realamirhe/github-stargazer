```python
from glob import glob
import pandas as pd
```

```python
datas = pd.concat(list(map(pd.read_csv, sorted(list(glob('./datas/*.csv'))))))
```

```python
datas.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
datas = datas.set_index('full_name')
```

```python
datas.info()
```

```python
import pandas.api.types as pat
features = {key: datas[key].nunique() for key in datas}

features
```

```python
categories = {
    f: "category"
    for f in features
    if pat.is_object_dtype(datas[f].dtypes) and features[f] < 10
}

categories
```

```python
datas = datas.astype(categories)
```

```python
datas.info()
```

```python
from sklearn import preprocessing

transformers = {c: preprocessing.LabelEncoder() for c in categories}
for c in transformers:
  datas[c] = transformers[c].fit_transform(datas[c])
# le.inverse_transform([0, 0, 1, 2])
```

```python
datas.info()
```

```python
integers = {
    key: datas[key].max() 
    for key in datas 
    if pat.is_integer_dtype(datas[key].dtype)
}

integers
```

```python
datas = datas.astype({
   i: 'int8' if integers[i] < 127 else 'int16' for i in integers 
})
```
